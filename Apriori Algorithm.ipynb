```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
with open('categories.txt', "r") as infile:
    transactions = [set(line.rstrip("\n").split(";")) for line in infile]
```

```python
to_file = []
for line in transactions:
    for element in line:
        to_file.append(element)
        
(unique, counts) = np.unique(to_file, return_counts=True)

joined = [{a:b} for a, b in zip(counts, unique)]
```

```python
with open('categories.txt', "r") as infile:
    transactions = [set(line.rstrip("\n").split(";")) for line in infile]

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
apriori(df, min_support=0.01, use_colnames = True)
```

```python
ap = apriori(df, min_support=0.01, use_colnames = True)
```

```python
ap = np.asarray(ap)
```
